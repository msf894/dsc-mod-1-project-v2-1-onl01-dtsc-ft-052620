```python
# Drops duplicate rows based on matching primary title and release date

movie_details_df = movie_details_df.drop_duplicates(subset=['primary_title',
                                                              'release_date',],
                                                      keep="first",
                                                      inplace=False)
movie_details_df.info()
```

```python
for col in movie_details_df:
    print(f'Viewing values in col: {col}')
    print(f'Top 5 values:\n{movie_details_df[col].value_counts()}')
```

### Data type conversion

```python
# Converts the columns containing dollar amounts into integers for easier
# comparisons later on

def convert_amt_to_int(df, col):
    df[col] = df[col].str.replace("$", "").str.replace(",", "").astype('int64')
    return df
```

```python
# Converts all the money columns into integers

money_cols = ['production_budget','domestic_gross','worldwide_gross']

for col in money_cols:
    movie_details_df = convert_amt_to_int(movie_details_df, col)
```

```python
#confirming data type changes

movie_details_df.info()
```

```python
# Combines domestic and worldwide gross into one summed column

movie_details_df['total_gross'] = movie_details_df['domestic_gross'] + movie_details_df['worldwide_gross']
movie_details_df.head()
```

```python
# Create a net profit column based on subtracting production budget from total gross

movie_details_df['net_profit'] = movie_details_df['total_gross'] - movie_details_df['production_budget']
movie_details_df.head()
```

```python
# Creates net profin margin column as a %

movie_details_df['net_profit_margin'] = round(((movie_details_df['net_profit'] / 
                                                movie_details_df['production_budget']) * 100),2)
movie_details_df.head()
```

```python
# function to convert the month number string to a integer

def monthToNum(shortMonth):
    months = {
        'Jan' : 1,
        'Feb' : 2,
        'Mar' : 3,
        'Apr' : 4,
        'May' : 5,
        'Jun' : 6,
        'Jul' : 7,
        'Aug' : 8,
        'Sep' : 9, 
        'Oct' : 10,
        'Nov' : 11,
        'Dec' : 12
        }
    return months[shortMonth]
```

```python
movie_details_df.head()
```

```python
# Apply the function to the column and confirmation of data type change

movie_details_df['release_month'] = movie_details_df['release_month'].apply(monthToNum)
movie_details_df.head()
```

### Removing unwanted columns

```python
# Removing excess columns that won't be used any longer

movie_details_df = movie_details_df.drop(columns = {'original_title','start_year', 'runtime_minutes','movie'})
movie_details_df.head()
```

### Splitting into genres

```python
# splits the genre string into a list

movie_details_df['genres'] = movie_details_df['genres'].apply(lambda x: x.split(",") if x else x)
movie_details_df.head()
```

```python
#generate a set of all unique genre names

all_genres = set()
for genres in movie_details_df['genres']:
    if genres:
        all_genres.update(genres)
```

```python
# creates columns with value 0 for each genre

for genre in all_genres:
    movie_details_df[genre] = np.zeros(shape=movie_details_df.shape[0])
                                
movie_details_df.head(15)
```

```python
# Changes value of column to 1 if name in genre column is found in genres

for index, row in movie_details_df.iterrows():
    if row['genres']:
        for genre in row['genres']:
            movie_details_df.loc[index, genre] = 1

movie_details_df.head(15)
```

```python
for col in movie_details_df:
    print(f'Viewing values in col: {col}')
    print(f'Top 5 values:\n{movie_details_df[col].value_counts()}')
```

We chose to specifically look at the median profit margin for each film genre as the key metric. The reason for this is that due to the massive fluctuations between total revenue and total expenses across filmmaking in general, the mean will be greatly inflated and won’t be an accurate measurement as a baseline

```python
genre_median = {}
for col in genre_cols:
    df_practice = movie_details_df_edited.loc[movie_details_df_edited[col] == 1.0] 
    genre_median[col] = round(df_practice['net_profit_margin'].median(),2)

genre_df = pd.DataFrame(list(genre_median.items()), columns = ['Genres','Median Profit Margin'])
genre_df = genre_df.sort_values('Median Profit Margin', ascending=False)
top10 = genre_df.head(10)
top10
```
